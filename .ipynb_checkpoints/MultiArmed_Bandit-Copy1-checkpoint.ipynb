```python
# import libraries
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import math
```

```python
class BanditData:
    def __init__(self, num, eps):
        self.eps = eps
        self.num = num
        self.overal_running_avg = 0
        self.overal_total_reward = 0
        
    def update_analysis(self, comBandit):
        if self.overal_running_avg < comBandit.running_avg:
            self.overal_running_avg = comBandit.running_avg
            
        if self.overal_total_reward < comBandit.total_reward:
            self.overal_total_reward = comBandit.total_reward
    @classmethod
    def output_cumulative_analysis(cs,data):
        print('Hello!')

class AdBandit:
    def __init__(self, ads):
        self.ads = ads
        self.running_avg = 0
        self.total_reward = 0
        self.times_played = 0
        
    def show(self,index):
        # return random results with normal distribution with variance of 1, using the average reward as the mean
        return self.ads[index]
    
    def update(self, nr):
        self.times_played += 1
        self.total_reward += nr
        self.running_avg += (nr - self.running_avg) / self.times_played
        
# create bandit class
class Bandit:
    def __init__(self, r):
        self.r_dist = r
        self.running_avg = 0
        self.total_reward = 0
        self.times_played = 0
        
    def play(self):
        # return random results with normal distribution with variance of 1, using the average reward as the mean
        return np.random.normal(self.r_dist, 1)
    
    def update(self, nr):
        self.times_played += 1
        self.total_reward += nr
        self.running_avg += (nr - self.running_avg) / self.times_played

# create environment where bandits are played 
class Environment:
    def __init__(self, eps, steps):
        self.steps = steps
        self.eps = eps
    
    def run(self, bandits): 
        total_mean = 0
        rewards = np.zeros(self.steps)
        for i in range(self.steps):             
            
            if np.random.rand() < self.eps:
                # explore, pick a random bandit
                index = np.random.choice(10)
            else:
                #exploit, choose bandit with best average outcome so far
                index = np.argmax([b.running_avg for b in bandits])
            
            selected = bandits[index]
            
            # get reward for bandit or advertisment
            if isinstance(selected,AdBandit):
                reward = selected.show(i)
            else:    
                reward = selected.play()
            
            # update the toal average
            total_mean = total_mean + (reward - total_mean) / (i+1)
            
            #update the average of the bandit we just played
            selected.update(reward)
            
            #append current total average to result array
            rewards[i] = total_mean
        return rewards, bandits

    def analyze(self, bandits):
        max_avg_index = np.argmax([b.running_avg for b in bandits])
        max_total_index = np.argmax([b.total_reward for b in bandits])
        return bandits[max_avg_index], bandits[max_total_index]
```

```python
# establish the number of arms, steps and tasks


def run_experiment(steps, tasks, func):
    # set up the reward arrays for each epsilon
    greedy_r = np.zeros(steps)
    ep01_r = np.zeros(steps)
    ep1_r = np.zeros(steps)
    
    # set up data analysis for each epsilon
    greedy_data = []
    ep1_data = []
    ep01_data = []
    
    # use the generation function to get the number of arms
    arms = len(func())

    # generate data classes
    for a in arms:
        greedy_data.append(a+1, BanditData('0.0'))
        
    for a in arms:
        ep1_data.append(a+1, BanditData('0.1'))
        
    for a in arms:
        ep01_data.append(a+1, BanditData('0.01'))

    for i in range(tasks):
        # create epsilon environments
        greedy = Environment(0.0, steps)
        e1 = Environment(0.1, steps)
        e01 = Environment(0.01, steps)
    

        # play each unique set of bandits and return the avg result
        # func provides the bandits we want to test
        # returns the rewards and the played bandits
        r_g, g_bandits = greedy.run(func())
        r_e1, e1_bandits = e1.run(func())
        r_e01, e01_bandits = e01.run(func())
        
        
        for k in arms:
            greedy_data[k].update_analysis(g_bandits[k])
            ep1_data[k].update_analysis(e1_bandits[k])
            ep01_data[k].update_analysis(e01_bandits[k])

        # add results to the cumulative average array
        greedy_r = greedy_r + (r_g - greedy_r) / (i + 1)
        ep01_r = ep01_r + (r_e01 - ep01_r) / (i + 1)
        ep1_r = ep1_r + (r_e1 - ep1_r) / (i + 1)
    return greedy_r, ep01_r, ep1_r, [greedy_data, ep1_data, ep01_data]
```

```python
def createGraphs(greedy, ep1, ep01, start, ranges):
    fig, axs = plt.subplots(2, 2,figsize=(10,10))
    fig.suptitle('10 arm bandit')

    for i in range(len(ranges)):
        r = ranges[i]
        a = axs[math.floor(i/2),i%2]
        a.plot(greedy[start:r], label="greedy")
        a.plot(ep1[start:r], label="epsilon=0.01")
        a.plot(ep01[start:r], label="epsilon=0.1")
        a.set_title("Average Reward Chart with " + str(r) + " tasks")
        
    fig.legend([axs[0, 0], axs[0, 1], axs[1,0], axs[1, 1]],
        labels=["greedy", "epsilon=0.01", "epsilon=0.1"],
        loc="center right",
        borderaxespad=0.1)
    plt.subplots_adjust(right=0.85)
    for ax in axs.flat:
        ax.set(xlabel='Steps', ylabel='Average Reward')
    plt.show()
```

```python
# create function to generate bandits for part 1
def generate_bandits():
    meansdist = np.random.normal(0, 1, 10)
    bandits = []
    for d in meansdist:
        new_bandit = Bandit(d)
        bandits.append(new_bandit)
    return bandits
    
```

```python
# run part 1
p1_g, p1_e1, p1_e01 = run_experiment(5000, 1000, generate_bandits)
```

```python
createGraphs(p1_g, p1_e1, p1_e01,0, [1000,3000,4000,5000])
```

```python
# load csv ad data for part 2
df = pd.read_csv('Ads_Optimisation.csv')
ad_array = df.to_numpy().transpose()
```

```python
# create function to generate ad bandits for part 2

def generate_ad_bandits():
    ad_bandits = []
    for a in ad_array:
        new_ad_bandit = AdBandit(a.tolist())
        ad_bandits.append(new_ad_bandit)
    return ad_bandits
```

```python
p2_g, p2_e1, p2_e01 = run_experiment(10000, 1000, generate_ad_bandits)
```

```python
createGraphs(p2_g, p2_e1, p2_e01, 100, [2000,3000,5000,10000])
```
