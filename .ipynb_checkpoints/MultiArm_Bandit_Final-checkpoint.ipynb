To run the experiment, we first create 2000 sets of 10 armed bandits. Each arm's true value is represented by a number randomly choosen nith normal distribution with mean 0 and vairance 1 (notated as N(0,1)). true_opt_arms stores which index has the largest true value. When an arm is 'pulled', the reward will be a randim number with the true value as the mean with variance 1, denoted as N(arm_true_value, 1). The estimated reward from each step is stored in Q and the bandit played is stored in N.
    We play the sets of bandits 1000, 3000, 5000, and ultimately 7500 times. The process for each pull has several steps. First we generate a random uniform number between 0 and 1. If the genrated number is less than the epsilon, we choose the bandit with the largest estimated reward so far. to determine whether the optimal bandit was the optimal bandit, we take the index selected and compare it to the index of true_opt_arms array for the given pull. If the indexes are the same, that means the optimal bandit was selected, and we increment the opt_arm_pull map by 1. Then, we retreive the reward from the bandit by the method described earlier, and record it in the rewards array for that pull. We update the current estimates similarly to how I described it in the notation. For meta data about the pulls, we track the running averages for each arm and the total rewards for each arm. We store these in an array of 10. 
    At the end of all the pulls, we get the average off the total rewards for the step, as well as calculate the max averages and max total for each arm. 

```python
import numpy as np
import matplotlib.pyplot as plt
import random
import pandas as pd 
```

```python
def bandit_strategy_1(n_bandit,k,eps, N, Q, true_opt_arms, opt_arm_pull, 
                      q_true, R_pull, bandit_plays,running_avg,running_max):
    
    for i in range(n_bandit):
        
        if random.random()<eps: 
            j=np.random.randint(k)
            
        else : 
            j=np.argmax(Q[i])

        if j==true_opt_arms[i] : # To calculate % optimal action
            opt_arm_pull['opt_arm_pull'] = opt_arm_pull['opt_arm_pull'] + 1

        reward=np.random.normal(q_true[i][j],1)
        R_pull.append(reward)
        N[i][j]=N[i][j]+1
        Q[i][j]=Q[i][j]+(reward-Q[i][j])/N[i][j]

        bandit_plays[j] += 1
        running_avg[j] = running_avg[j] + reward
        running_max[j] = running_max[j] + reward
    for i in range(k):
        running_avg[i] = running_avg[i]/n_bandit
```

```python
def run_experiment(eps, steps, n_bandit, strat):
    print(f'running for eps {eps}')
    
    k=10 # number of arms in each bandit problem

    q_true=np.random.normal(0,1,(n_bandit,k)) # generating {n_bandit} bandit sets of {k} bandits
    true_opt_arms=np.argmax(q_true,1) # the index of max value of each set to determine optimal bandit percentage

    Q=np.zeros((n_bandit,k)) # reward estimated
    N=np.ones((n_bandit,k)) # number of times each arm was pulled # each arm is pulled atleast once

    Qi=np.random.normal(q_true,1) # initial pulling of all arms

    R_eps=[]
    R_eps.append(0)
    R_eps.append(np.mean(Qi))
    R_eps_opt = []
    
    m_running_avg = np.zeros(k) # keep track of running averages for each arm
    m_running_max = np.zeros(k) # max reward recorded for each arm


    for pull in range(steps) :  
        R_pull=[] # all rewards in this pull/time-step
        opt_arm_pull = {'opt_arm_pull':0}  # number of times the best arm out of the 10 was selected
        bandit_plays = np.zeros(k) # number of times each bandit was played
        running_avg = np.zeros(k) # keep track of running averages for each arm
        running_max = np.zeros(k) # max reward recorded for each arm
        
        # bandit strategy callback
        strat(n_bandit,k,eps, N, Q, true_opt_arms, opt_arm_pull, 
              q_true, R_pull, bandit_plays,running_avg,running_max)

            
        for i in range(10):
            if m_running_max[i] < running_max[i]:
                m_running_max[i] = running_max[i]
            if m_running_avg[i] < running_avg[i]:
                m_running_avg[i] = running_avg[i]

        avg_R_pull=np.mean(R_pull)
        R_eps.append(avg_R_pull)
        R_eps_opt.append(float(opt_arm_pull['opt_arm_pull'])*100/2000)
        
    max_index = np.argmax(m_running_max)
    avg_index = np.argmax(m_running_avg)
    arm_data = {
        'eps':eps,
        'steps': steps,
        'max_total': {'max': round(m_running_max[max_index], 2), 'arm': max_index+1},
        'max_avg': {'max': round(m_running_avg[avg_index], 2), 'arm': avg_index+1}
    }
        
    return R_eps,R_eps_opt, arm_data
```

```python

def run_experiments_for_epsilons(steps, strat):
    print(f'running experiments for {steps}')
    g_a, g_o, g_data = run_experiment(0.0, steps, 2000, strat)
    e1_a, e1_o, e1_data = run_experiment(0.1, steps, 2000, strat)
    e01_a, e01_o, e01_data = run_experiment(0.01, steps, 2000, strat)
    
    return {
        'greedy': {
            'rewards': g_a,
            'optimize':g_o,
            'arm_data':g_data
        },
        'e1': {
            'rewards': e1_a,
            'optimize':e1_o,
            'arm_data':e1_data
        },
        'e01': {
            'rewards': e01_a,
            'optimize':e01_o,
            'arm_data':e01_data
        }
    }
```

```python
def print_bandit_data(data):
    eps = data["eps"]
    m_t = data['max_total']['max']
    m_t_arm = data['max_total']['arm']
    m_a = data['max_avg']['max']
    m_a_arm = data['max_avg']['arm']
    print(f'For eps  {eps}')
    print(f'Max Total: with {m_t} from arm {m_t_arm}')
    print(f'Max Avg: with {m_a} from arm {m_a_arm}')
    
def print_data_group(d_array):
    steps = d_array[0]['steps']
    print(f'For {steps} steps')
    for d in d_array:
        print_bandit_data(d)
        print("\n")
    print('--------------------\n')

def createGraphs(greedy, ep1, ep01, start, r, title):
    fig, a = plt.subplots(figsize=(10,10))
    fig.suptitle(f"10 arm bandit {r} steps")

    a.plot(greedy[start:r], label="greedy")
    a.plot(ep1[start:r], label="epsilon=0.1")
    a.plot(ep01[start:r], label="epsilon=0.01")
    a.set_title(f"{title} " + str(r) + " tasks")
    fig.legend(
        labels=["greedy", "epsilon=0.1", "epsilon=0.01"],
        loc="center right",
        borderaxespad=0.1)
    plt.show()
```

```python
data_1000 = run_experiments_for_epsilons(1000, bandit_strategy_1)
data_3000 = run_experiments_for_epsilons(3000, bandit_strategy_1)
data_5000 = run_experiments_for_epsilons(5000, bandit_strategy_1)
data_7500 = run_experiments_for_epsilons(7500, bandit_strategy_1)
print('done')
```

```python
# show graph for avg reward
createGraphs(
    data_7500['greedy']['rewards'],
    data_7500['e1']['rewards'],
    data_7500['e01']['rewards'],
    0,
    len(data_7500['greedy']['rewards']),
    'Average Reward for ')

# show graph for percent of optimal bandit picked
createGraphs(
    data_7500['greedy']['optimize'],
    data_7500['e1']['optimize'],
    data_7500['e01']['optimize'],
    0,
    len(data_7500['greedy']['optimize']),
    'Percentage that Optimal Bandit was picked for')

for data in [data_1000,data_3000,data_5000,data_7500]:
    g_data = data['greedy']['arm_data']
    e1_data = data['e1']['arm_data']
    e01_data = data['e01']['arm_data']
    print_data_group([g_data, e1_data, e01_data])
```
