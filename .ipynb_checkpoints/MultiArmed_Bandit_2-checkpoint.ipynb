```python
import numpy as np
import pandas as pd
import random
import math
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
class Bandit:
    def __init__(self, r):
        self.mean_r = 0
        self.times_played = 0
        self.reward_dist = r
        
    def play(self):
        return np.random.normal(self.reward_dist, 1)
    
    def update(self, r):
        self.times_played += 1
        self.mean_r = self.mean_r + ((r - self.mean_r) / self.times_played)

def play_bandits(bandits, eps, steps): 
    
    rewards = np.zeros(steps)
    total_mean = 0
    for s in range(steps):
        p = np.random.rand()             
        if p < eps:
            index = np.random.choice(len(bandits))
        else:
            index = np.argmax([b.mean_r for b in bandits]) 

        selected_bandit = bandits[index]

        reward = selected_bandit.play()
        selected_bandit.update(reward)

        total_mean = total_mean + ((reward - total_mean) / (s+1))
        rewards[s] = total_mean
    return rewards

def run_experiment(bandits, iterations,eps,steps):
    result_array = np.zeros(steps)
    for x in range(iterations):
        rs = play_bandits(bandits,eps,steps)
        result_array = result_array + (rs-result_array)/(x+1)
    return result_array

if __name__ == '__main__': 
    dists = np.random.normal(0, 1, 10)
    bandits = []
    for d in dists:
        new_bandit = Bandit(d)
        bandits.append(new_bandit)

    
    iterations = 2000
    steps = 1000
    results = []
    eps_list = [0.0,0.1, 0.1]
#     c_1 = run_experiment(bandits, iterations, 0.0, steps)
    c_2 = play_bandits(bandits, 0.01, steps)
    plt.figure(figsize=(10,8))
#     plt.plot(c_1, label = '0.0')
    plt.plot(c_2, label = '0.01')

    plt.legend() 
    
```

```python
class bandit:
    def __init__(self, arms, type, steps):
#         self.n = 0
        self.steps = steps
        self.arms = arms
        self.type = type
        self.mean_r = 0

        self.arm_n = np.zeros(arms)
        self.reward = np.zeros(steps)
        self.arm_r = np.zeros(arms)
        # unknown mean reward of a action
        self.meansdist = np.random.normal(0, 1, arms)

    def run(self): 
        for i in range(self.steps):
            # find random numbers and determine arm choice
            rn = np.random.rand()             
            if rn < self.type:
                act = np.random.choice(self.arms)
            else:
                act = np.argmax(self.arm_r)

            # increment counters
#             self.n += 1
            self.arm_n[act] += 1
            
            # determine reward from Gaussian distribution with q*(a) as mean as varaiance 1
            reward = np.random.normal(self.meansdist[act], 1)
            # determine the mean reward and reward for the arm chosen
            self.mean_r = self.mean_r + (reward - self.mean_r) / (i+1)
            self.arm_r[act] = self.arm_r[act] + (reward - self.arm_r[act]) / self.arm_n[act]
            # update the reward array for each step
            self.reward[i] = self.mean_r
```

```python
# establish the number of arms, steps and tasks
arms = 10
steps = 2000
tasks = 2000

# set up the long-term reward arrays for each type
greedy_r = np.zeros(steps)
ep01_r = np.zeros(steps)
ep1_r = np.zeros(steps)

for i in range(tasks):
    # initialize each type
    greedy = bandit(arms, 0, steps)
    e01 = bandit(arms, 0.01, steps)
    e1 = bandit(arms, 0.1, steps)

    # run learning action (pull arm) this is done the number of steps times
    greedy.run()
    e01.run()
    e1.run()

    # deteremine the long-term rewards
    greedy_r = greedy_r + (greedy.reward - greedy_r) / (i + 1)
#     ep01_r = ep01_r + (e01.reward - ep01_r) / (i + 1)
#     ep1_r = ep1_r + (e1.reward - ep1_r) / (i + 1)

# draw the average reward output chart
print(greedy_r[500:510])
plt.figure(figsize=(10,8))
plt.plot(greedy_r, label="greedy")
plt.plot(ep01_r, label="epsilon=0.01")
plt.plot(ep1_r, label="epsilon=0.1")
plt.legend(bbox_to_anchor=(1, 0.5))
plt.xlabel("Steps")
plt.ylabel("Average Reward")
plt.title("Average Reward Chart with " + str(tasks) + " tasks")
plt.show()
```

### Part 1: Creating general 10 armed bandit experiment

```python
# created with help from https://www.geeksforgeeks.org/epsilon-greedy-algorithm-in-reinforcement-learning/
class Bandit: 
    def __init__(self, r): 
        self.reward = r 
        self.running_avg = 0
        self.N = 0

    # Play bandit, get reward
    def play(self):  
        return np.random.normal(self.reward,1)
  
  # Update the running average reward of the bandit
    def update(self, r): 
        self.N += 1        
        self.running_avg = self.running_avg + (r-self.running_avg)/self.N
  
  
def run_experiment(bandit_rewards, eps, N): 
    bandits = []
    
    for br in bandit_rewards:
        bandits.append(Bandit(br))
        
    rewards = np.empty(N) 
    
    mean = 0
    for i in range(N): 
        p = np.random.rand()
        if p<eps: 
            #explore
            b_index = np.random.choice(len(bandits)) 
        else: 
            #exploit. Choose the bandit with the best return thus far
            b_index = np.argmax([b.running_avg for b in bandits]) 

        selected_bandit = bandits[b_index]
        #play selected bandit and update running average
        r = selected_bandit.play() 
        selected_bandit.update(r) 
        
        mean = mean + (r-mean)/steps
        
        #append result to array
        rewards[i] = mean 
    
    # create a cumulative sum of the average rewards
#     cumulative_average = np.cumsum(data) / (np.arange(N) + 1) 
    return rewards 

def get_avg_experiment(steps,array,eps,N):
    avg_array = np.zeros(N)
    for i in range(steps):
        rs = run_experiment(array,eps, N) 
        avg_array = avg_array + (rs - avg_array)/(i+1)
    return avg_array

if __name__ == '__main__': 
    reward_array = np.random.normal(0, 1, 10)
    random.shuffle(reward_array)
    N = 1000
    s = 2000
    c_1 = get_avg_experiment(s,reward_array,0.1, N)
    c_2 = get_avg_experiment(s,reward_array,0.01, N)
    c_3 = get_avg_experiment(s,reward_array,0.0, N)
    plt.figure(figsize=(10,8))
    plt.plot(c_1, label = '0.1')
    plt.plot(c_2, label = '0.01')
    plt.plot(c_3, label = '0.0')
    plt.legend() 
```

### Part 2: Using general bandit experiment on advertising data

```python
# load csv data
df = pd.read_csv('Ads_Optimisation.csv')
ad_array = df.to_numpy().transpose()
```

```python
import math
class Advertisment: 
    def __init__(self, a): 
        self.ad_array = a
        self.ads_length = len(a)
        self.running_avg = 0
        self.N = 0

    # Show ad
    def show(self):  
        ad_index = math.floor(random.uniform(0,1) * self.ads_length)
        return self.ad_array[ad_index]
  
  # Update the running average reward of the ad
    def update(self, r): 
        self.N += 1
        self.running_avg = self.running_avg + ((1/self.N)*(r-self.running_avg))
        
        
def run_experiment(advertisments, eps, N): 
        
    data = np.empty(N) 
    
    eps_mod = N+1 if eps == 0.0 else N/(N*eps)

    for i in range(N): 
    
        if i%eps_mod == 0: 
            #explore
            a_index = np.random.choice(len(advertisments)) 
        else: 
            #exploit. Choose the ad with the best return thus far
            a_index = np.argmax([a.running_avg for a in advertisments]) 
        
        #play selected ad and update running average
        r = advertisments[a_index].show() 
        advertisments[a_index].update(r) 
        
        #append result to array
        data[i] = r
    
    # create a cumulative sum of the average rewards
    cumulative_average = np.cumsum(data) / (np.arange(N) + 1) 
    return cumulative_average 

def get_avg_experiment(steps,array,eps,N):
    avg_array = np.zeros(N)
    for i in range(steps):
        c = run_experiment(array,eps, N) 
        avg_array = np.add(avg_array,c)
    return avg_array/steps



if __name__ == '__main__': 
    advertisments = []
    
    for ad in ad_array:
        newAd = Advertisment(ad)
        advertisments.append(newAd)
    random.shuffle(advertisments)
        
    N = 2000
    s = 2000
    c_1 = get_avg_experiment(s,advertisments,0.1, N) 
    c_2 = get_avg_experiment(s,advertisments,0.01, N) 
    c_3 = get_avg_experiment(s,advertisments,0.0, N) 

    
    plt.plot(c_1, label = '0.1')
    plt.plot(c_2, label = '0.01')
    plt.plot(c_3, label = '0.0')
    plt.xscale('log')
    plt.legend() 
```
